Punto 8 Dado el array [3, 8, 15, 20, 7, 12, 18, 1], obtener un nuevo array con los valores mayores a 10

punto 10Crear una matriz 3x3 con números aleatorios entre 1 y 10. Calcular la suma de cada fila y de cada
columna.


```python

arreglo = np.random.choice(10, size=9, replace=False)
matriz= arreglo.reshape(3,3)



print(matriz)

suma_filas = np.sum(matriz, axis=1)

# Suma de cada columna (axis=0)
suma_columnas = np.sum(matriz, axis=0)

print("\nSuma de cada fila:", suma_filas)
print("Suma de cada columna:", suma_columnas)
```

 Dado el array [2,5,8,11,14,17], reemplazar los valores mayores a 10 por -1.

12. Crear un array con números del 1 al 30 y obtener un nuevo array que contenga solo los números
pares

```python

arreglo = np.arange(1, 31)

pares = arreglo[arreglo % 2 == 0]

print("arreglo original:", arreglo)
print("Números pares:", pares)
```

13. Crear dos matrices A (2x3) y B (3x2) y calcular su producto matricial.

```python

A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = np.array([[7, 8],
              [9, 10],
              [11, 12]])

C = np.dot(A, B)  

print("Matriz A:")
print(A)

print("\nMatriz B:")
print(B)

print("\nProducto A x B:")
print(C)
```

15. Dado el array [10,20,30,40,50], normalizar los valores usando (valor - minimo) / (maximo - minimo)

```python


array = np.array([10,20,30,40,50])

minimo = np.min(array)
maximo = np.max(array)

normalizado = (array - minimo) / (maximo - minimo)

print(normalizado)
```
