```python
import pandas as pd

# Crear un DataFrame (como una tabla)
datos = {
    "Nombre": ["Juan", "Ana", "Luis"],
    "Edad": [20, 25, 30],
    "Ciudad": ["Córdoba", "Buenos Aires", "Rosario"]
}

df = pd.DataFrame(datos)

# Mostrar la tabla
print("DataFrame completo:")
print(df)

# Mostrar solo una columna
print("\nEdades:")
print(df["Edad"])

# Promedio de edades
print("\nPromedio de edad:")
print(df["Edad"].mean())
```
