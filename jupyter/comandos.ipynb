# EJEMPLO 9
- Sumar dos arreglos unidimensionales utilizando linspace
- concatenar los dos arreglos
- Ordenar los valores del nuevo array

```python
a = np.linspace(10, 20, 6)
b = np.linspace(5, 25, 6)
suma = a + b
print(a)
print(b)
print(suma)
```

# EJEMPLO 12
- Buscar el min, max, promedio, desvio estandar y suma de una lista random de enteros entre 0 y 20 de 10 numeros

```python
q = np.random.random_integers(20, size=10)
print(min(q))
print(max(q))
print(q.mean())
print(q.std())
print(q.sum())
```

# EJEMPLO 12
- Encontrar el min y el max en un arrglo bidimensional de 5x4 con numeros del 0 al 20

```python
p = np.random.random_integers(20, size=(5,4))
print(p)
print(p.mean())
print(p.min(axis=0))
print(p.max(axis=1))
print(p.min())
print(p.max())
```

# EJEMPLO 14
- Apilamiento vertical y horizontal

```python
u = np.random.random_integers(20, size=(5,4))
print(u)
v = np.random.random_integers(20, size=(5,4))
print(v)
print(np.vstack((u,v)))
print(np.hstack((u,v)))
```
