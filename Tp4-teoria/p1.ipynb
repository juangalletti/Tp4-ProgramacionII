```python
import pandas as pd

# Leer los archivos
df1 = pd.read_csv("argentina_cars1.csv")
df2 = pd.read_csv("argentina_cars2.csv")

# Concatenar por filas
df_total = pd.concat([df1, df2], axis=0)

# Opcional: resetear índices
df_total = df_total.reset_index(drop=True)

df_total.to_csv("ConcatenadoCars1&cars2.csv")#guardoelarchivo concatenado en csv


filtro = df_total[(df_total['money'] > 3000000) &
                  (df_total['money'] < 5000000)]
print(filtro)
```

4- Graficar un histograma con Matplotlib 
Descripción: Leer el archivo mundial.csv, armar un dataframe con selecciones y cantidad de 
mundiales ganados. Luego graficar un histograma.

```python
import matplotlib.pyplot as plt

df = pd.read_csv("mundial.csv")
print(df.head())  # muestra las primeras filas


data = {
    "Seleccion": ["Brasil", "Alemania", "Uruguay","Francia","Argentina","Italia","inglaterra","España"],
    "Mundiales": [5, 4, 4, 3, 3, 4, 1, 1]
}

Mundiales = pd.DataFrame(data)

print(Mundiales)


plt.bar(Mundiales['Seleccion'], Mundiales['Mundiales'])
plt.title('Mundiales por selección')
plt.xlabel('Selección')
plt.ylabel('Cantidad de copas')
plt.xticks(rotation=45)
plt.show()



```
