matplotlib o seaborn.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("ConcatenadoCars1&cars2.csv")

# Limpiar nombres por las dudas
df.columns = df.columns.str.strip()

fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# 🔹 1. Top marcas
df['brand'].value_counts().head(10).plot(kind='bar', ax=axs[0,0])
axs[0,0].set_title("Top 10 marcas")
axs[0,0].set_xlabel("Marca")
axs[0,0].set_ylabel("Cantidad")

# 🔹 2. Distribución de precios
df['money'].plot(kind='hist', bins=20, ax=axs[0,1])
axs[0,1].set_title("Distribución de precios")
axs[0,1].set_xlabel("Precio")

# 🔹 3. Autos por año
df['year'].value_counts().sort_index().plot(ax=axs[1,0])
axs[1,0].set_title("Autos por año")
axs[1,0].set_xlabel("Año")
axs[1,0].set_ylabel("Cantidad")

# 🔹 4. Precio promedio por marca
df.groupby('brand')['money'].mean().sort_values(ascending=False).head(10).plot(kind='bar', ax=axs[1,1])
axs[1,1].set_title("Precio promedio por marca")
axs[1,1].set_xlabel("Marca")
axs[1,1].set_ylabel("Precio promedio")

plt.tight_layout()
plt.show()
```
